### **Mirai Botnet (2016)**
- **Attribution:** The initial Mirai botnet attack wasn't state-sponsored but laid the groundwork for subsequent attacks possibly linked to nation-state actors.
- **Description:** Mirai compromised IoT devices like IP cameras and home routers, turning them into a massive botnet that launched distributed denial-of-service (DDoS) attacks. Variants of Mirai have since been used in more sophisticated operations, potentially linked to nation-state actors.
- **Exploitation:**
    - Variants of the Mirai botnet continued to exploit IoT devices with outdated firmware that had not been patched. This includes devices with vulnerabilities that had long been identified and should have been addressed by firmware updates.
- **Impact:** It disrupted major internet services and highlighted the vulnerabilities in IoT devices. Legacy IoT devices continued to be vulnerable, leading to repeated DDoS attacks and further botnet expansions.
- **Vulnerabilities & Exploits:**
- Mirai exploited weak or default credentials on IoT devices, such as hardcoded usernames and passwords, to gain control.
- **CVEs:**
    - The original Mirai botnet did not exploit specific CVEs but rather relied on brute-force attacks using a list of default credentials found on IoT devices.

### **VPNFilter (2018)**
- **Attribution:** Linked to the Russian group Fancy Bear (APT28).
- **Description:** This malware targeted routers and network-attached storage (NAS) devices, exploiting known vulnerabilities to gain control and extract information, modify traffic, and even render devices inoperable.
- **Exploitation:**
    - VPNFilter specifically targeted routers and NAS devices with outdated firmware. The malware was able to exploit known vulnerabilities that had not been patched, allowing the attackers to gain persistent access to the devices.
- **Impact:** Over 500,000 devices in more than 50 countries were affected, highlighting the risks posed by vulnerable IoT devices.
- **Vulnerabilities & Exploits:**
- VPNFilter targeted vulnerabilities in routers and NAS devices, exploiting known flaws in older firmware versions.
- **CVEs:**
    - **CVE-2018-14847**: MikroTik RouterOS before 6.42.1 had a directory traversal vulnerability that allowed reading of files outside the root directory.
    - **CVE-2017-12074**: A command injection vulnerability in Linksys routers.
    - **CVE-2018-10561** and **CVE-2018-10562**: Remote code execution vulnerabilities in Dasan GPON routers.
    - **CVE-2017-8418**: Remote code execution vulnerability in TP-Link routers.

### **Triton/Trisis (2017)**
- **Attribution:** Believed to be linked to a Russian state-sponsored group.
- **Description:** This attack targeted Industrial Control Systems (ICS) in critical infrastructure, specifically safety systems, potentially using compromised IoT devices as entry points. The goal was to cause physical damage to the plant.
- **Exploitation:**
    - Triton targeted outdated industrial control systems, including safety instrumented systems (SIS) that were not regularly updated. The attackers took advantage of legacy firmware and insufficient patching, allowing them to install malware that could cause physical damage.
- **Impact:** It marked a significant escalation in the use of cyber-physical attacks on critical infrastructure. The attack highlighted the dangers of unpatched legacy systems in critical infrastructure, especially those connected to IoT devices.
- **Vulnerabilities & Exploits:**
- Triton exploited vulnerabilities in Schneider Electric’s Triconex safety controllers, specifically targeting the software used for industrial control.
- **CVEs:**
    - **CVE-2017-7905**: An authenticated remote code execution vulnerability in Schneider Electric's Triconex Tricon system due to insufficient input validation in its TriStation protocol.
    - **CVE-2017-7921**: Vulnerability in Triconex software that could allow for unauthorized access and potentially malicious firmware uploads.

### **Iranian Cyber Attacks on Water Systems (2020)**
- **Attribution:** Iranian state-sponsored actors.
- **Description:** These attacks targeted IoT-connected water infrastructure, attempting to manipulate water levels and flow in Israel. The attack aimed to disrupt civilian life and critical infrastructure.
- **Exploitation:**
    - These attacks targeted SCADA systems and other critical infrastructure, many of which included legacy IoT devices. The attackers exploited unpatched vulnerabilities in these older systems, which were not regularly updated, to gain unauthorized access.
- **Impact:** While the attack was mitigated, it underscored the vulnerabilities in IoT devices within critical infrastructure. The attacks demonstrated how legacy systems in critical infrastructure, including IoT devices, could be a weak point when not properly maintained.
- **Vulnerabilities & Exploits:**
- Specific details of the CVEs exploited in these attacks are not publicly disclosed. The attack likely exploited weaknesses in the IoT devices and SCADA systems controlling water infrastructure.
- **CVEs:**
    - No specific CVEs have been publicly attributed to this attack, but vulnerabilities in outdated SCADA systems and IoT devices were likely exploited.

## Commonalities Among Nation-State Attacks Against IoT Vulnerabilities
### 1. **Weak or Default Credentials**
   - **Example:** Mirai Botnet exploited IoT devices by using a brute-force attack against default or weak credentials.
   - **Commonality:** Many IoT devices are shipped with default usernames and passwords that users often do not change, making them easy targets for attackers.

### 2. **Outdated Firmware and Software**
   - **Example:** VPNFilter exploited known vulnerabilities in routers and NAS devices, many of which had outdated firmware.
   - **Commonality:** IoT devices are often not updated regularly with the latest security patches, leaving them vulnerable to exploits that have already been publicly disclosed.

### 3. **Inadequate Input Validation and Lack of Security Controls**
   - **Example:** Triton/Trisis exploited vulnerabilities in industrial control systems where inadequate input validation allowed unauthorized access.
   - **Commonality:** IoT devices often lack robust security controls, such as input validation and proper authentication mechanisms, making them susceptible to exploitation.

### 4. **Supply Chain Vulnerabilities**
   - **Example:** Operation ShadowHammer targeted the ASUS Live Update Utility, compromising the software supply chain and impacting IoT devices that used the affected software.
   - **Commonality:** Attackers can target the supply chain of IoT devices to introduce malicious code before the devices even reach the consumer, exploiting trust in the supply chain.

### 5. **Exploitation of Protocol Vulnerabilities**
   - **Example:** Ripple20 vulnerabilities in the Treck TCP/IP stack allowed attackers to perform remote code execution and denial of service attacks.
   - **Commonality:** Many IoT devices use lightweight and sometimes insecure communication protocols, which can be exploited to gain unauthorized access or cause disruptions.

### 6. **Remote Code Execution (RCE) Vulnerabilities**
   - **Example:** Multiple vulnerabilities exploited in these attacks, such as CVE-2020-11896 in Ripple20, allowed for remote code execution.
   - **Commonality:** RCE vulnerabilities are particularly dangerous in IoT devices as they allow attackers to execute arbitrary code remotely, potentially gaining full control over the device.

### 7. **Lack of Segmentation in Networks**
   - **Example:** The Iranian cyberattacks on water systems likely exploited IoT devices that were part of inadequately segmented networks, allowing attackers to pivot from IoT devices to more critical infrastructure.
   - **Commonality:** IoT devices are often part of larger networks without sufficient segmentation, allowing attackers to use them as entry points to access more sensitive areas of the network.

### 8. **Insufficient Security Monitoring and Detection**
   - **Example:** VPNFilter and Triton/Trisis attacks were not immediately detected, allowing the attackers to remain undetected for extended periods.
   - **Commonality:** Many IoT devices lack the ability to log and monitor security events, making it difficult to detect and respond to intrusions promptly.

### 9. **Exploitability Due to Widespread Use and Lack of Standardization**
   - **Example:** Ripple20 vulnerabilities were found in a widely used TCP/IP stack, affecting millions of devices.
   - **Commonality:** The widespread use of certain software components in IoT devices, coupled with a lack of standardized security practices across different manufacturers, makes these devices a broad attack surface.
